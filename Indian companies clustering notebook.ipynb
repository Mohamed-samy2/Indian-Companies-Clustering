Displaying the distribution plot for each individual column to visualize the spread and pattern of values within each attribute of the dataset.

```python
for col in data.columns:
    if data[col].dtype !='object':
        sns.displot(x=data[col])
    plt.show()
```

- The majority of columns exhibit a right-skewed distribution, indicating a concentration of values towards lower ranges with a tail extending towards higher values, as previously identified.
- The "Starting Year" column, while representing a left-skewed distribution due to earlier years, is less relevant for focus in this analysis.

```python
def show_distributions(data,col):
    plt.figure(figsize=(8, 4))
    # Plotting the first distribution
    plt.subplot(1, 3, 1)
    sns.histplot(np.log1p(data[col]),kde=True,  color='skyblue')
    plt.title('log transformation')


    # Plotting the second distribution
    plt.subplot(1, 3, 2)
    sns.histplot(np.cbrt(data[col]),kde=True,  color='salmon')
    plt.title('cubic root transformation')

    # Plotting the third distribution
    plt.subplot(1, 3, 3)
    sns.histplot(np.sqrt(data[col]), kde=True, color='green')
    plt.title('sqaure root transformation')

    plt.suptitle(f'Distributions for column {col}', fontsize=16, y=1.05)


    plt.tight_layout()
    plt.show()
    print(f"Skewness with log = {np.log1p(data[col]).skew()} and with cubic root = {np.cbrt(data[col]).skew()} and with square root = {np.sqrt(data[col]).skew()}")
```

Displaying the distributions after applying different transformations (e.g., log transformation, square root transformation) to the data columns and calculating the skewness values to assess the changes in the skewness of each attribute.



```python
columns=['Funding Round','Funding Amount in $','No. of Investors']
for col in columns:
    show_distributions(data,col)
```

- Log transformation yields the most optimal results for reducing skewness in the "Funding Round" and "Funding Amount in $" columns, resulting in the lowest skewness values among the applied transformations.
- Among the transformations applied, the cubic root transformation demonstrates the most effective skewness reduction for the "Number of Investors" column, resulting in the minimum skewness value observed.

```python
def show_qq(data,col):
    plt.figure(figsize=(8, 4))
    # Plotting the first distribution
    plt.subplot(1, 3, 1)
    stats.probplot(np.log1p(data[col]), dist="norm", plot=plt)
    plt.title('log transformation')
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Sample Quantiles')



    # Plotting the second distribution
    plt.subplot(1, 3, 2)
    stats.probplot(np.cbrt(data[col]), dist="norm", plot=plt)
    plt.title('cubic root transformation')
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Sample Quantiles')


    # Plotting the third distribution
    plt.subplot(1, 3, 3)
    stats.probplot(np.sqrt(data[col]), dist="norm", plot=plt)
    plt.title('sqaure root transformation')
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Sample Quantiles')


    plt.suptitle(f'Q-Q plot for column {col}', fontsize=16, y=1.05)


    plt.tight_layout()
    plt.show()
```

Generating Q-Q plots for each transformation applied to visualize the degree of conformity to normal distribution and assess the effectiveness of the transformation in reducing skewness.

```python
for col in columns:
    show_qq(data,col)
```

- The Q-Q plots validate our previous assertions, showcasing the effectiveness of the transformationssuch as log transformation for "Funding Round" and "Funding Amount in $" columns, and cubic root transformation for "Number of Investors" columnin reducing skewness and aligning the distributions closer to a normal distribution.

```python
def calculate_outliers(col):
    sorted(col)
    Q1, Q3 = col.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

Calculating and presenting the count and percentage of outliers within each column to demonstrate the extent of extreme values present in the dataset.

```python
lower_limit, upper_limit = calculate_outliers((data['Funding Round']))
lower_outliers=len(data[(data['Funding Round'])<lower_limit])
upper_outliers=len(data[(data['Funding Round'])>upper_limit])
print(f"total outliers in col Funding Round  = ",upper_outliers+lower_outliers," with precentage of ",(upper_outliers+lower_outliers)/data.shape[0]*100)
```

```python
lower_limit, upper_limit = calculate_outliers((data['Funding Amount in $']))
lower_outliers=len(data[(data['Funding Amount in $'])<lower_limit])
upper_outliers=len(data[(data['Funding Amount in $'])>upper_limit])
print(f"total outliers in col Funding Amount in $  = ",upper_outliers+lower_outliers," with precentage of ",(upper_outliers+lower_outliers)/data.shape[0]*100)
```

```python
lower_limit, upper_limit = calculate_outliers((data['No. of Investors']))
lower_outliers=len(data[(data['No. of Investors'])<lower_limit])
upper_outliers=len(data[(data['No. of Investors'])>upper_limit])
print(f"total outliers in col No. of Investors  = ",upper_outliers+lower_outliers," with precentage of ",(upper_outliers+lower_outliers)/data.shape[0]*100)
```

Executing the specified transformations individually for each column within the dataset to mitigate skewness and normalize the distribution of values across the attributes.


```python
data['Funding Round']=np.log1p(data['Funding Round'])
data['Funding Amount in $']=np.log1p(data['Funding Amount in $'])
data['No. of Investors']=np.cbrt(data['No. of Investors'])
```

### 3.1) KMeans

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import rand_score
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
```

Obtaining and specifying the features from the preprocessed and scaled dataset that will be used for the clustering process, ensuring relevant attributes are selected for the analysis.

```python
data_scaled=data_scaled.drop(columns=['Company','City','Starting Year','No. of Employees'])
data=data.drop(columns=['Company','City','Starting Year'])
```

```python
data.head()
```

Instantiate the KMeans algorithm with a range of 'K' values to identify the optimal number of clusters by employing the elbow method for clustering analysis.

```python
K=range(1,15)
clusters = []

for k in K:
    kmeans = KMeans(n_clusters=k, init='random', max_iter=1000000,random_state=42)
    kmeans.fit_predict(data_scaled)
    clusters.append(kmeans.inertia_)
```

Generating an elbow plot by varying the number of clusters (K) to visualize the within-cluster sum of squares (WCSS) against different values of K, aiding in the selection of the optimal number of clusters for the KMeans algorithm.

```python
plt.figure(figsize=(10,5))
plt.xlabel('Number of clusters (K)')
plt.ylabel('Cost')
plt.title("Elbow method")
plt.grid()
plt.plot(K,clusters,'o-')
```

Using a visualizer such as the "Yellowbrick" library to assist in determining the optimal number of clusters (K) by plotting an elbow graph, displaying the within-cluster sum of squares (WCSS) against different values of K for enhanced clarity and decision-making in cluster selection.

```python
from yellowbrick.cluster import KElbowVisualizer
kmeans=KMeans(init='random', max_iter=1000000,random_state=42)
visualizer=KElbowVisualizer(kmeans,k=(1,15))
visualizer.fit(data_scaled)
visualizer.poof()
plt.show()
```

- The analysis suggests that the optimal number of clusters (K) for the dataset is three (k=3) based on the elbow method, indicating that partitioning the data into three clusters provides an appropriate balance between complexity and interpretability for this specific dataset.

Calculate both the Silhouette score and Davies-Bouldin index to evaluate the quality and coherence of the clustering results obtained with the chosen number of clusters (k=3), providing insights into the clustering performance.

```python
silhoutte_list=[]
davies_list=[]
k_list=[]
for k in range(2,15):
  kmeans = KMeans(n_clusters=k, init='random', max_iter=1000000,random_state=42)
  model= kmeans.fit_predict(data_scaled)
  cluster_labels = kmeans.labels_
  silhoutte_list.append(np.round(silhouette_score(data_scaled,cluster_labels),4))
  davies_list.append(np.round(davies_bouldin_score(data_scaled,cluster_labels),4))
  k_list.append(k)
  # print(f'At {k} Clusters the Silhouette Score is {np.round(silhouette_score(data,cluster_labels),3)} and Davies Bouldin Score is {np.round(davies_bouldin_score(data,cluster_labels),3)} ')


evaluation_metric=pd.DataFrame({'number of clusters':k_list,'Silhouette Score':silhoutte_list,'Davies Bouldin Score':davies_list})
```

```python
plt.figure(figsize=(8, 5))
plt.plot(evaluation_metric['number of clusters'], evaluation_metric['Silhouette Score'], marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different k values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(evaluation_metric['number of clusters'], evaluation_metric['Davies Bouldin Score'], marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Davies Bouldin Score')
plt.title('Davies Bouldin Score for Different k values')
plt.show()
```

### 3.2) DBSCAN.

```python
from sklearn.cluster import DBSCAN
from sklearn.model_selection import ParameterGrid
```

Performing an iterative process by trying various epsilon and minimum samples combinations within the DBSCAN algorithm to determine the optimal parameters based on the silhouette score for clustering analysis.

```python
param_grid = {
    'eps': [0.1,18,19,20, 0.2, 0.3,0.9,1,0.96,0.65,0.25,1,2,3,1.5,5,20,30,5,6,8,9,4,1.9,1.8,2.5,2.6,2.4,3,3.5,3.9,0.01,0.001],
    'min_samples': np.arange(2,10)
}

best_silhouette_score = -1  # Initialize with a lower score
  # Initialize with a higher score
best_params_silhouette = None


# Iterate through parameter combinations
for params in ParameterGrid(param_grid):
    dbscan = DBSCAN(**params)
    dbscan.fit(data_scaled)

    # Calculate silhouette score for current parameters
    labels = dbscan.labels_
    if len(set(labels)) > 1:  # Silhouette score requires at least 2 clusters
        silhouette = silhouette_score(data_scaled, labels)

        # Update best parameters if the silhouette score is higher
        if silhouette > best_silhouette_score:
            best_silhouette_score = silhouette
            best_params_silhouette = params






print("Best Parameters for Silhouette Score:", best_params_silhouette)
print("Best Silhouette Score:", best_silhouette_score)
```

Creating an instance of the DBSCAN model with the specified parameters, setting epsilon (eps) to 0.9 and the minimum number of samples to 9, to conduct the clustering analysis using these parameter values.

```python
dbscan = DBSCAN(eps = 0.9, min_samples=9).fit(data_scaled)
db_labels = dbscan.labels_
data["Cluster"]=db_labels
```

Compute both the Silhouette score and Davies-Bouldin index to assess the quality and coherence of the clusters generated by the DBSCAN model with the given parameter settings.

```python
print("Silhouette Score For DBSCAN is ",np.round(silhouette_score(data_scaled,db_labels),4))
print("Davies Bouldin Score For DBSCAN is ",np.round(davies_bouldin_score(data_scaled,db_labels),4))
```

### 3.3) Hierarchical clustering.

Enhancing the model by employing the Ward linkage method with the Euclidean metric for hierarchical clustering, allowing for the creation of clusters based on the specified linkage criterion and distance metric for improved clustering performance.

```python
from scipy.cluster.hierarchy import linkage,fcluster
from scipy.cluster.hierarchy import dendrogram

# Use the linkage() function
distance_matrix = linkage(data_scaled, method='ward', metric='euclidean')
```

Generating a dendrogram visualization to explore the hierarchical structure and determine the optimal number of clusters by observing the variations in cluster distances, aiding in the selection of an appropriate number of clusters for hierarchical clustering.

```python
dn = dendrogram(distance_matrix)
plt.figure(figsize=(15,10))
# Display the dendogram
plt.show()
```

The dendrogram visualization indicates that three clusters appear to be the optimal choice based on the observed cluster distances and structure, suggesting that partitioning the data into three clusters provides a suitable representation for this hierarchical clustering analysis.

```python
hierarchy_labels = fcluster(distance_matrix,3,criterion='maxclust')
```

```python
print("Silhouette Score For hierarchical clustering is ",np.round(silhouette_score(data_scaled,hierarchy_labels),4))
print("Davies Bouldin Score For hierarchical clustering is ",np.round(davies_bouldin_score(data_scaled,hierarchy_labels),4))
```

```python
data["Cluster"]=hierarchy_labels
```
